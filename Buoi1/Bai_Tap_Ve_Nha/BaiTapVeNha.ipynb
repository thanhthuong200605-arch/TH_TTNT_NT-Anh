```python
#Bài 1:
def tron_mang(a, b):
    result = []
    n = min(len(a), len(b))

    # Lấy min từng cặp
    for i in range(n):
        result.append(min(a[i], b[i]))

    # Thêm phần còn lại
    if len(a) > len(b):
        result.extend(a[n:])
    else:
        result.extend(b[n:])

    return result


# Test
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

print(tron_mang(a, b))  # [2, 7, 1, 3, 2, 8]
```

```python
# ====== NHẬP MA TRẬN ======
def nhap_ma_tran(n, m):
    a = []
    for i in range(n):
        row = list(map(int, input(f"Nhap dong {i+1}: ").split()))
        a.append(row)
    return a


def xuat_ma_tran(a):
    for row in a:
        print(*row)


# ====== a. TỔNG TAM GIÁC TRÊN CHÉO PHỤ ======
def tong_tam_giac_tren_cheo_phu(a):
    n = len(a)
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong


# ====== b. ĐỔI ÂM → TRỊ TUYỆT ĐỐI ======
def doi_am(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])


# ====== c. THAY SỐ CHẴN BẰNG x ======
def thay_chan(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x


# ====== d. KIỂM TRA TOÀN CHẴN ======
def toan_chan(a):
    for row in a:
        for val in row:
            if val % 2 != 0:
                return False
    return True


# ====== e. MA TRẬN ĐỐI XỨNG ======
def doi_xung(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return False
    return True


# ====== f. CHÉO CHÍNH TĂNG DẦN ======
def cheo_chinh_tang(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][i] >= a[i + 1][i + 1]:
            return False
    return True


# ====== g. XUẤT TAM GIÁC DƯỚI CHÉO PHỤ ======
def xuat_tam_giac_duoi_cheo_phu(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end=" ")
        print()


# ====== h. CHÉO PHỤ GIẢM DẦN ======
def cheo_phu_giam(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][n - i - 1] <= a[i + 1][n - i - 2]:
            return False
    return True


# ====== CHẠY CHƯƠNG TRÌNH ======
n = int(input("Nhap n (ma tran vuong): "))
a = nhap_ma_tran(n, n)

print("\nMa tran vua nhap:")
xuat_ma_tran(a)

print("\nTong tam giac tren cheo phu:", tong_tam_giac_tren_cheo_phu(a))

doi_am(a)
print("\nSau khi doi so am:")
xuat_ma_tran(a)

x = int(input("\nNhap x de thay so chan: "))
thay_chan(a, x)
print("Sau khi thay so chan:")
xuat_ma_tran(a)

print("\nToan chan?", toan_chan(a))
print("Doi xung?", doi_xung(a))
print("Cheo chinh tang?", cheo_chinh_tang(a))

print("\nTam giac duoi cheo phu:")
xuat_tam_giac_duoi_cheo_phu(a)

print("\nCheo phu giam?", cheo_phu_giam(a))
```

```python
#Bài 3:
# ====== NHẬP DỮ LIỆU ======
def nhap_sv():
    ten = input("Nhap ten: ")
    tuoi = int(input("Nhap tuoi: "))
    diem = float(input("Nhap diem: "))
    return {"ten": ten, "tuoi": tuoi, "diem": diem}


data = {}

# Nhập 3 sinh viên
for i in range(1, 4):
    print(f"\nNhap sinh vien {i}:")
    data[f"sv{i}"] = nhap_sv()


# ====== XUẤT DỮ LIỆU ======
print("\nDanh sach sinh vien:")
for ma, info in data.items():
    print(f"{ma}: Ten={info['ten']}, Tuoi={info['tuoi']}, Diem={info['diem']}")


# ====== TRUY CẬP 1 PHẦN TỬ ======
print("\nThong tin sv1:")
print(data["sv1"])


# ====== SỬA DỮ LIỆU ======
data["sv1"]["diem"] = 10
print("\nSau khi sua diem sv1:")
print(data["sv1"])


# ====== THÊM SINH VIÊN ======
print("\nThem sinh vien moi:")
data["sv4"] = nhap_sv()


# ====== XOÁ SINH VIÊN ======
xoa = input("\nNhap ma sinh vien can xoa (vd: sv2): ")
if xoa in data:
    del data[xoa]
    print("Da xoa!")
else:
    print("Khong ton tai!")


# ====== DUYỆT LẠI DANH SÁCH ======
print("\nDanh sach sau cung:")
for ma, info in data.items():
    print(f"{ma}: Ten={info['ten']}, Tuoi={info['tuoi']}, Diem={info['diem']}")
```

```python
#Bài 4:
# ====== GHI FILE ======
def ghi_file(ten_file):
    with open(ten_file, "w", encoding="utf-8") as f:
        n = int(input("Nhap so dong: "))
        for i in range(n):
            line = input(f"Nhap dong {i+1}: ")
            f.write(line + "\n")
    print("Da ghi file!")


# ====== ĐỌC FILE ======
def doc_file(ten_file):
    try:
        with open(ten_file, "r", encoding="utf-8") as f:
            print("\nNoi dung file:")
            print(f.read())
    except FileNotFoundError:
        print("File khong ton tai!")


# ====== ĐỌC TỪNG DÒNG ======
def doc_tung_dong(ten_file):
    try:
        with open(ten_file, "r", encoding="utf-8") as f:
            print("\nDoc tung dong:")
            for line in f:
                print(line.strip())
    except FileNotFoundError:
        print("File khong ton tai!")


# ====== THÊM NỘI DUNG (APPEND) ======
def them_file(ten_file):
    with open(ten_file, "a", encoding="utf-8") as f:
        line = input("Nhap noi dung them: ")
        f.write(line + "\n")
    print("Da them vao file!")


# ====== CHẠY CHƯƠNG TRÌNH ======
ten_file = "data.txt"

while True:
    print("\n===== MENU =====")
    print("1. Ghi file")
    print("2. Doc file")
    print("3. Doc tung dong")
    print("4. Them vao file")
    print("0. Thoat")

    chon = input("Chon: ")

    if chon == "1":
        ghi_file(ten_file)
    elif chon == "2":
        doc_file(ten_file)
    elif chon == "3":
        doc_tung_dong(ten_file)
    elif chon == "4":
        them_file(ten_file)
    elif chon == "0":
        break
    else:
        print("Chon sai!")
```

```python
#Bài 5:
# ====== IMPORT THƯ VIỆN ======
import pandas as pd
import matplotlib.pyplot as plt


# ====== NHẬP DỮ LIỆU ======
def nhap_du_lieu():
    ds_ten = []
    ds_diem = []

    n = int(input("Nhap so sinh vien: "))
    for i in range(n):
        ten = input(f"Nhap ten sv {i+1}: ")
        diem = float(input("Nhap diem: "))
        ds_ten.append(ten)
        ds_diem.append(diem)

    return ds_ten, ds_diem


# ====== TẠO DATAFRAME ======
def tao_dataframe(ten, diem):
    df = pd.DataFrame({
        "Ten": ten,
        "Diem": diem
    })
    return df


# ====== HIỂN THỊ DỮ LIỆU ======
def hien_thi(df):
    print("\nDanh sach sinh vien:")
    print(df)


# ====== TÍNH TOÁN ======
def thong_ke(df):
    print("\nThong ke:")
    print("Diem trung binh:", df["Diem"].mean())
    print("Diem cao nhat:", df["Diem"].max())
    print("Diem thap nhat:", df["Diem"].min())


# ====== LỌC SINH VIÊN ======
def loc_sinh_vien(df):
    print("\nSinh vien diem >= 5:")
    print(df[df["Diem"] >= 5])


# ====== VẼ BIỂU ĐỒ ======
def ve_bieu_do(df):
    plt.bar(df["Ten"], df["Diem"])
    plt.title("Diem sinh vien")
    plt.xlabel("Ten")
    plt.ylabel("Diem")
    plt.show()


# ====== LƯU FILE CSV ======
def luu_file(df):
    df.to_csv("sinhvien.csv", index=False)
    print("Da luu file sinhvien.csv")


# ====== CHẠY CHƯƠNG TRÌNH ======
ten, diem = nhap_du_lieu()

df = tao_dataframe(ten, diem)

hien_thi(df)
thong_ke(df)
loc_sinh_vien(df)
ve_bieu_do(df)
luu_file(df)
```
