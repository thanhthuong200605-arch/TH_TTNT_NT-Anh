```python
#Bai 13:
def tron_mang_cong(mang1, mang2):
    ket_qua = []
    do_dai1 = len(mang1)
    do_dai2 = len(mang2)
    do_dai_nho = min(do_dai1, do_dai2)

    # Cộng các phần tử tương ứng
    for i in range(do_dai_nho):
        ket_qua.append(mang1[i] + mang2[i])

    # Thêm phần còn lại
    if do_dai1 > do_dai_nho:
        ket_qua.extend(mang1[do_dai_nho:])
    elif do_dai2 > do_dai_nho:
        ket_qua.extend(mang2[do_dai_nho:])

    return ket_qua


# Test 1
mang_a = [3, 9, 1, 4]
mang_b = [2, 7, 4, 3, 2, 8]

ket_qua = tron_mang_cong(mang_a, mang_b)

print("Mảng a:", mang_a)
print("Mảng b:", mang_b)
print("Mảng kết quả:", ket_qua)


# Test 2
mang_x = [1, 2, 3, 4, 5]
mang_y = [10, 20]

ket_qua_xy = tron_mang_cong(mang_x, mang_y)

print("\nMảng x:", mang_x)
print("Mảng y:", mang_y)
print("Mảng kết quả (x,y):", ket_qua_xy)
```

```python
#Bai 14:
import random

# a. Tạo mảng ngẫu nhiên
def tao_mang(m, n):
    return [[random.randint(1, 20) for _ in range(n)] for _ in range(m)]

# b. Xuất dòng k
def xuat_dong(a, k):
    print("Dòng", k, ":", a[k])

# c. Xuất cột k
def xuat_cot(a, k):
    print("Cột", k, ":", [row[k] for row in a])

# d. Tìm dòng có tổng > 45 và lớn nhất
def dong_tong_lon_nhat(a):
    max_sum = -1
    index = -1
    for i, row in enumerate(a):
        s = sum(row)
        if s > 45 and s > max_sum:
            max_sum = s
            index = i
    return index, max_sum

# e. Tìm cột có tích nhỏ nhất
def cot_tich_nho_nhat(a):
    n = len(a[0])
    min_prod = float('inf')
    index = -1

    for j in range(n):
        prod = 1
        for i in range(len(a)):
            prod *= a[i][j]
        if prod < min_prod:
            min_prod = prod
            index = j

    return index, min_prod

# f. Phần tử dòng chẵn, cột lẻ
def phan_tu_chan_le(a):
    result = []
    for i in range(len(a)):
        if i % 2 == 0:
            for j in range(len(a[0])):
                if j % 2 == 1:
                    result.append(a[i][j])
    return result

# g. TBC phần tử chẵn thuộc dòng lẻ
def tbc_chan_dong_le(a):
    total = 0
    count = 0

    for i in range(len(a)):
        if i % 2 == 1:
            for x in a[i]:
                if x % 2 == 0:
                    total += x
                    count += 1

    return total / count if count != 0 else 0

# h. TBC phần tử biên
def tbc_bien(a):
    m, n = len(a), len(a[0])
    total = 0
    count = 0

    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                total += a[i][j]
                count += 1

    return total / count

# i. Tinh trung binh tich phan tu khong thuoc bien
def tbc_khong_bien(a):
    m, n = len(a), len(a[0])
    total = 0
    count = 0

    for i in range(1, m-1):
        for j in range(1, n-1):
            total += a[i][j]
            count += 1

    return total / count if count != 0 else 0


# ================= TEST =================
m, n = 4, 5
a = tao_mang(m, n)

print("Ma trận:")
for row in a:
    print(row)

xuat_dong(a, 1)
xuat_cot(a, 2)

print("d: Dòng tổng >45 lớn nhất:", dong_tong_lon_nhat(a))
print("e: Cột tích nhỏ nhất:", cot_tich_nho_nhat(a))
print("f: Phần tử dòng chẵn cột lẻ:", phan_tu_chan_le(a))
print("g: TBC chẵn dòng lẻ:", tbc_chan_dong_le(a))
print("h: TBC phần tử biên:", tbc_bien(a))
print("i: TB tich phan tu khong thuoc bien:", tbc_khong_bien(a))
```

```python
#Bai 15:
# --- CÂU A:
class SinhVien:
    def __init__(self, ma_sv, ten, nam_sinh, diem_tb):
        self.ma_sv = ma_sv
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.diem_tb = diem_tb

# --- CÂU B:
danh_sach_sv = [
    SinhVien("01DH000001", "Phan Anh Tuấn", 2006, 8.5),
    SinhVien("02DH000002", "Lê Thị Lan", 2004, 4.2),
    SinhVien("03CD000001", "Nguyễn Văn Lan", 2006, 7.0),
    SinhVien("04DH000003", "Phan Thị Bưởi", 2000, 9.0),
    SinhVien("05DH000004", "Trần Hoà", 2005, 5.5)
]

# Câu c: Đếm sinh viên đủ điều kiện lên lớp (DTB >= 5)
def cau_c(ds):
    count = sum(1 for sv in ds if sv.diem_tb >= 5)
    print(f"c. Số sinh viên đủ điều kiện lên lớp: {count}")

# Câu d: Xuất các sinh viên đủ 20 tuổi (Tính tại năm 2026)
def cau_d(ds):
    nam_hien_tai = 2026
    print(f"d. Danh sách sinh viên đủ 20 tuổi (sinh năm {nam_hien_tai - 20}):")
    co_sv = False
    for sv in ds:
        if nam_hien_tai - sv.nam_sinh == 20:
            print(f"   - Tên: {sv.ten} | Mã SV: {sv.ma_sv}")
            co_sv = True
    if not co_sv:
        print("   - Không có sinh viên nào đủ 20 tuổi.")

# Câu e: Đếm sinh viên hệ Đại học (Mã có 'DH' ở vị trí index 2,3)
def cau_e(ds):
    count = sum(1 for sv in ds if len(sv.ma_sv) >= 4 and sv.ma_sv[2:4] == "DH")
    print(f"e. Số sinh viên học hệ Đại học (có 'DH' ở vị trí 2,3): {count}")

# Câu f: Đếm sinh viên có tên "Lan"
def cau_f(ds):
    count = 0
    for sv in ds:
        ten_tach_ra = sv.ten.split()
        if ten_tach_ra and ten_tach_ra[-1] == "Lan":
            count += 1
    print(f"f. Số sinh viên có tên 'Lan': {count}")

# Câu g: Đếm sinh viên có họ "Phan"
def cau_g(ds):
    count = sum(1 for sv in ds if sv.ten.startswith("Phan"))
    print(f"g. Số sinh viên có họ 'Phan': {count}")

if __name__ == "__main__":
    print("="*40)
    print("      KẾT QUẢ QUẢN LÝ SINH VIÊN 2026")
    print("="*40)

    cau_c(danh_sach_sv)
    print("-" * 40)
    cau_d(danh_sach_sv)
    print("-" * 40)
    cau_e(danh_sach_sv)
    print("-" * 40)
    cau_f(danh_sach_sv)
    print("-" * 40)
    cau_g(danh_sach_sv)
    print("="*40)
```
