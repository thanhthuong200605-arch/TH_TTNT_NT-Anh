```python
#Bài 5: Khung sườn thuật toán Alpha-Beta cho Cờ Vua (Chess)
import math
import copy
import numpy # Import numpy để hiển thị bàn cờ đẹp hơn

# --- Biểu diễn bàn cờ và các quân cờ (Rất đơn giản, chỉ mang tính minh họa) ---
# Trong thực tế, đây sẽ là một lớp (class) Board phức tạp hơn nhiều

# Giả sử bàn cờ là ma trận 8x8
KICH_THUOC_BAN_CO = 8

# Định nghĩa các quân cờ (đơn giản hóa)
GIA_TRI_QUAN_CO = {
    'P': 1, 'N': 3, 'B': 3, 'R': 5, 'Q': 9, 'K': 100,  # Trắng
    'p': -1, 'n': -3, 'b': -3, 'r': -5, 'q': -9, 'k': -100 # Đen
} # Giá trị quân cờ

def tao_ban_co_ban_dau():
    # Đây là một biểu diễn bàn cờ vua rất cơ bản.
    # Một implementation thực tế sẽ phức tạp hơn nhiều với các quân cờ và vị trí cụ thể.
    bang = [
        ['r', 'n', 'b', 'q', 'k', 'b', 'n', 'r'],
        ['p', 'p', 'p', 'p', 'p', 'p', 'p', 'p'],
        [None, None, None, None, None, None, None, None],
        [None, None, None, None, None, None, None, None],
        [None, None, None, None, None, None, None, None],
        [None, None, None, None, None, None, None, None],
        ['P', 'P', 'P', 'P', 'P', 'P', 'P', 'P'],
        ['R', 'N', 'B', 'Q', 'K', 'B', 'N', 'R']
    ]
    return bang

def lay_nguoi_choi_hien_tai(bang, dem_luot):
    # Giả sử 'X' là Trắng, 'O' là Đen
    # Trong cờ vua, Trắng đi trước.
    return 'X' if dem_luot % 2 == 0 else 'O'

# --- Các hàm cần thiết cho Cờ Vua (chỉ là placeholders) ---
def tao_cac_nuoc_di(bang, nguoi_choi):
    # Đây là hàm phức tạp nhất trong Cờ Vua.
    # Nó phải tạo ra TẤT CẢ các nước đi hợp lệ cho người chơi hiện tại.
    # Ví dụ: di chuyển tốt, mã, tượng, xe, hậu, vua, nhập thành, bắt tốt qua đường, ...
    # Trả về một danh sách các nước đi có thể, ví dụ: [((hang1, cot1), (hang2, cot2)), ...]
    # Với Alpha-beta, việc sắp xếp nước đi (move ordering) hiệu quả sẽ cải thiện đáng kể tốc độ cắt tỉa.
    # Đối với ví dụ này, chúng ta sẽ trả về một số nước đi đơn giản (tiến 1 ô)
    cac_nuoc_di_kha_thi = []
    for hang in range(KICH_THUOC_BAN_CO):
        for cot in range(KICH_THUOC_BAN_CO):
            quan_co = bang[hang][cot]
            if quan_co is not None and \
               ((nguoi_choi == 'X' and quan_co.isupper()) or \
                (nguoi_choi == 'O' and quan_co.islower())):

                # Ví dụ đơn giản: quân tốt di chuyển một ô thẳng
                if quan_co.lower() == 'p': # Chỉ áp dụng cho tốt
                    if nguoi_choi == 'X': # Tốt Trắng đi lên
                        if hang - 1 >= 0 and bang[hang - 1][cot] is None:
                            cac_nuoc_di_kha_thi.append(((hang, cot), (hang - 1, cot)))
                    else: # Tốt Đen đi xuống
                        if hang + 1 < KICH_THUOC_BAN_CO and bang[hang + 1][cot] is None:
                            cac_nuoc_di_kha_thi.append(((hang, cot), (hang + 1, cot)))
                # Thêm các luật di chuyển cho các quân khác và các nước đi ăn quân ở đây

    if not cac_nuoc_di_kha_thi:
        # Nếu không có nước đi nào được tạo (ví dụ đơn giản không bao phủ hết), tạo một nước đi giả
        # Để tránh lỗi khi không tìm thấy nước đi nào hợp lệ theo luật đơn giản
        # Trong một AI cờ vua thực thụ, điều này có nghĩa là game kết thúc (chiếu hết hoặc hòa)
        for r in range(KICH_THUOC_BAN_CO):
            for c in range(KICH_THUOC_BAN_CO):
                if bang[r][c] is None:
                    return [((r,c), (r,c))] # Trả về một nước đi giả đến ô trống đầu tiên
        return [((0,0), (0,0))] # Fallback dummy move if no empty cell

    return cac_nuoc_di_kha_thi

def thuc_hien_nuoc_di(bang, nuoc_di):
    # Thực hiện một nước đi trên bàn cờ. Trả về bàn cờ mới.
    # Trong cờ vua, nước đi bao gồm di chuyển quân, ăn quân, nhập thành, phong cấp.
    ban_co_moi = copy.deepcopy(bang)
    (tu_hang, tu_cot), (den_hang, den_cot) = nuoc_di

    # Kiểm tra xem đây có phải là nước đi 'giả' khi không có nước đi hợp lệ nào không
    if (tu_hang, tu_cot) == (den_hang, den_cot) and ban_co_moi[tu_hang][tu_cot] is None:
        return ban_co_moi # Không làm gì nếu là nước đi giả đến ô trống

    ban_co_moi[den_hang][den_cot] = ban_co_moi[tu_hang][tu_cot]
    ban_co_moi[tu_hang][tu_cot] = None
    return ban_co_moi

def hoan_tac_nuoc_di(bang, nuoc_di):
    # Hoàn tác một nước đi. Cần thiết cho việc tìm kiếm sâu để quay lại trạng thái trước.
    # Trong cờ vua, phức tạp hơn vì phải lưu trữ quân bị ăn, trạng thái nhập thành, ...
    # Đối với ví dụ này, chúng ta sẽ không thực hiện nó vì thuc_hien_nuoc_di đã trả về deepcopy.
    pass

def danh_gia_ban_co(bang, nguoi_choi):
    # Đây là hàm "trái tim" của AI Cờ Vua. Nó gán một giá trị số cho trạng thái bàn cờ.
    # Giá trị dương có lợi cho Trắng, âm có lợi cho Đen.
    # Trong thực tế, hàm này sẽ xem xét: giá trị quân cờ, vị trí quân, cấu trúc tốt,
    # an toàn của vua, kiểm soát trung tâm, khả năng tấn công/phòng thủ, nước chiếu, ...
    diem = 0
    for hang in range(KICH_THUOC_BAN_CO):
        for cot in range(KICH_THUOC_BAN_CO):
            quan_co = bang[hang][cot]
            if quan_co in GIA_TRI_QUAN_CO:
                diem += GIA_TRI_QUAN_CO[quan_co]
    return diem

def tro_choi_ket_thuc(bang):
    # Kiểm tra xem trò chơi đã kết thúc chưa (chiếu hết, hòa, hết nước đi).
    # Rất phức tạp trong Cờ Vua.
    # Đối với ví dụ này, chúng ta sẽ đơn giản hóa.
    # Để minh họa, game sẽ kết thúc nếu không có nước đi hợp lệ nào được tạo ra bởi generate_moves
    # Hoặc nếu một bên hết quân chính (ví dụ, vua bị ăn - điều này sẽ phức tạp hơn để kiểm tra)
    # Vì generate_moves trả về dummy move khi không có nước đi, ta có thể dựa vào đó.
    nguoi_choi_x_co_the_di_chuyen = tao_cac_nuoc_di(bang, 'X')
    nguoi_choi_o_co_the_di_chuyen = tao_cac_nuoc_di(bang, 'O')

    # Nếu cả hai bên chỉ có thể thực hiện nước đi dummy, coi như hết nước đi (hòa hoặc chiếu hết tùy ngữ cảnh)
    if nguoi_choi_x_co_the_di_chuyen == [((0,0),(0,0))] and nguoi_choi_o_co_the_di_chuyen == [((0,0),(0,0))]:
         return True

    return False # Giả sử game chưa kết thúc để minh họa minimax

def la_chieu_het(bang, nguoi_choi):
    # Kiểm tra chiếu hết
    return False

def la_hoa_co(bang, nguoi_choi):
    # Kiểm tra hòa (hết nước đi)
    return False

# --- Thuật toán Minimax với Cắt tỉa Alpha-beta ---
def minimax_alpha_beta(bang, do_sau, alpha, beta, nguoi_choi_toi_da_hoa, dem_luot):
    if do_sau == 0 or tro_choi_ket_thuc(bang):
        return danh_gia_ban_co(bang, 'X' if nguoi_choi_toi_da_hoa else 'O'), None

    nguoi_choi_hien_tai = lay_nguoi_choi_hien_tai(bang, dem_luot)
    nuoc_di_tot_nhat = None

    if nguoi_choi_toi_da_hoa: # Người chơi tối đa hóa (Trắng - 'X')
        danh_gia_toi_da = -math.inf
        cac_nuoc_di = tao_cac_nuoc_di(bang, nguoi_choi_hien_tai)
        for nuoc_di in cac_nuoc_di:
            ban_co_moi = thuc_hien_nuoc_di(bang, nuoc_di)
            gia_tri_danh_gia, _ = minimax_alpha_beta(ban_co_moi, do_sau - 1, alpha, beta, False, dem_luot + 1)
            if gia_tri_danh_gia > danh_gia_toi_da:
                danh_gia_toi_da = gia_tri_danh_gia
                nuoc_di_tot_nhat = nuoc_di
            alpha = max(alpha, gia_tri_danh_gia)
            if beta <= alpha:
                break  # Cắt tỉa Beta
        return danh_gia_toi_da, nuoc_di_tot_nhat
    else: # Người chơi tối thiểu hóa (Đen - 'O')
        danh_gia_toi_thieu = math.inf
        cac_nuoc_di = tao_cac_nuoc_di(bang, nguoi_choi_hien_tai)
        for nuoc_di in cac_nuoc_di:
            ban_co_moi = thuc_hien_nuoc_di(bang, nuoc_di)
            gia_tri_danh_gia, _ = minimax_alpha_beta(ban_co_moi, do_sau - 1, alpha, beta, True, dem_luot + 1)
            if gia_tri_danh_gia < danh_gia_toi_thieu:
                danh_gia_toi_thieu = gia_tri_danh_gia
                nuoc_di_tot_nhat = nuoc_di
            beta = min(beta, gia_tri_danh_gia)
            if beta <= alpha:
                break  # Cắt tỉa Alpha
        return danh_gia_toi_thieu, nuoc_di_tot_nhat

def tim_nuoc_di_tot_nhat(bang, do_sau, nguoi_choi, dem_luot):
    if nguoi_choi == 'X': # Trắng là người tối đa hóa
        _, nuoc_di_tot_nhat = minimax_alpha_beta(bang, do_sau, -math.inf, math.inf, True, dem_luot)
    else: # Đen là người tối thiểu hóa
        _, nuoc_di_tot_nhat = minimax_alpha_beta(bang, do_sau, -math.inf, math.inf, False, dem_luot)
    return nuoc_di_tot_nhat

# --- Hàm chạy chính (Ví dụ) ---
if __name__ == "__main__":
    ban_co_hien_tai = tao_ban_co_ban_dau()
    dem_luot = 0
    DO_SAU_TOI_DA = 2 # Độ sâu tìm kiếm nhỏ để ví dụ không quá chậm

    print("--- Bàn cờ ban đầu (Đơn giản hóa) ---")
    for hang in ban_co_hien_tai:
        print(hang)

    # Chọn bên chơi
    while True:
        chon_nguoi_choi = input("Bạn muốn chơi quân 'X' (Trắng) hay 'O' (Đen)? ").upper()
        if chon_nguoi_choi in ['X', 'O']:
            nguoi_choi_cua_ban = chon_nguoi_choi
            may = 'O' if nguoi_choi_cua_ban == 'X' else 'X'
            break
        else:
            print("Lựa chọn không hợp lệ. Vui lòng nhập 'X' hoặc 'O'.")

    print(f"Bạn chọn chơi quân: {nguoi_choi_cua_ban}")
    print(f"Máy chơi quân: {may}")

    while not tro_choi_ket_thuc(ban_co_hien_tai): # Game chạy đến khi kết thúc
        nguoi_choi_hien_tai = lay_nguoi_choi_hien_tai(ban_co_hien_tai, dem_luot)
        print(f"\n--- Lượt của người chơi: {nguoi_choi_hien_tai} ---")

        if nguoi_choi_hien_tai == nguoi_choi_cua_ban: # Lượt của người chơi
            print("--- Bàn cờ hiện tại ---")
            print(numpy.array(ban_co_hien_tai))
            while True:
                try:
                    print(f"Bạn là {nguoi_choi_cua_ban}. Nhập nước đi của bạn (tu_hang tu_cot den_hang den_cot, ví dụ: 6 0 5 0): ")
                    input_str = input().split()
                    if len(input_str) != 4:
                        raise ValueError("Vui lòng nhập đủ 4 số.")

                    tu_hang, tu_cot, den_hang, den_cot = map(int, input_str)
                    nuoc_di_cua_ban = ((tu_hang, tu_cot), (den_hang, den_cot))

                    # Kiểm tra nước đi hợp lệ
                    cac_nuoc_di_hop_le = tao_cac_nuoc_di(ban_co_hien_tai, nguoi_choi_cua_ban)

                    # Loại bỏ nước đi dummy nếu có
                    cac_nuoc_di_hop_le = [m for m in cac_nuoc_di_hop_le if m != ((0,0),(0,0)) and not (m[0]==m[1] and ban_co_hien_tai[m[0][0]][m[0][1]] is None)]

                    if not (0 <= tu_hang < KICH_THUOC_BAN_CO and \
                            0 <= tu_cot < KICH_THUOC_BAN_CO and \
                            0 <= den_hang < KICH_THUOC_BAN_CO and \
                            0 <= den_cot < KICH_THUOC_BAN_CO):
                        print("Lỗi: Tọa độ nằm ngoài bàn cờ. Vui lòng nhập lại.")
                    elif nuoc_di_cua_ban not in cac_nuoc_di_hop_le:
                        print(f"Lỗi: Nước đi {nuoc_di_cua_ban} không hợp lệ cho quân {nguoi_choi_cua_ban}. Vui lòng nhập lại.")
                        print(f"Các nước đi hợp lệ (chỉ tốt tiến 1 ô, không ăn quân): {cac_nuoc_di_hop_le}")
                    else:
                        ban_co_hien_tai = thuc_hien_nuoc_di(ban_co_hien_tai, nuoc_di_cua_ban)
                        break
                except ValueError as e:
                    print(f"Lỗi: Đầu vào không hợp lệ. {e}. Vui lòng nhập lại.")
                except Exception as e:
                    print(f"Có lỗi xảy ra: {e}. Vui lòng nhập lại.")
        else: # Lượt của máy
            print(f"Máy ({may}) đang tính toán nước đi...")
            nuoc_di_tot_nhat = tim_nuoc_di_tot_nhat(ban_co_hien_tai, DO_SAU_TOI_DA, may, dem_luot)

            if nuoc_di_tot_nhat:
                print(f"Máy đã đi: {nuoc_di_tot_nhat}")
                ban_co_hien_tai = thuc_hien_nuoc_di(ban_co_hien_tai, nuoc_di_tot_nhat)
                print("--- Bàn cờ sau nước đi của máy ---")
                print(numpy.array(ban_co_hien_tai))
            else:
                print("Máy không tìm thấy nước đi hợp lệ.")
                break

        dem_luot += 1

    print("\n--- Trò chơi kết thúc ---")
    # Bạn có thể thêm kiểm tra người thắng cuộc hoặc hòa ở đây dựa trên hàm danh_gia_ban_co
    diem_cuoi = danh_gia_ban_co(ban_co_hien_tai, nguoi_choi_cua_ban) # Đánh giá từ góc nhìn của người chơi
    if diem_cuoi > 0: # Trắng có lợi thế
        if nguoi_choi_cua_ban == 'X':
            print("Bạn (X) có vẻ đang thắng!")
        else:
            print("Máy (X) có vẻ đang thắng!")
    elif diem_cuoi < 0: # Đen có lợi thế
        if nguoi_choi_cua_ban == 'O':
            print("Bạn (O) có vẻ đang thắng!")
        else:
            print("Máy (O) có vẻ đang thắng!")
    else:
        print("Trò chơi kết thúc với kết quả hòa.")

    print("Lưu ý: Để có một AI cờ vua thực thụ, các hàm tao_cac_nuoc_di, thuc_hien_nuoc_di, hoan_tac_nuoc_di và danh_gia_ban_co cần được triển khai rất chi tiết và chính xác theo luật cờ vua.")
```

```python
#Bài 6:
import tkinter as tk
from tkinter import messagebox, simpledialog

class CaroNxN:
    def __init__(self, n=3):
        self.window = tk.Tk()
        self.window.title(f"Caro {n}x{n}")
        self.window.configure(bg="#f5f5f5")

        self.n = n
        # Xác định số ô cần để thắng
        self.so_o_thang = 3 if n <= 5 else 5

        self.nguoi_choi_hien_tai = "X"
        self.bang = [["" for _ in range(n)] for _ in range(n)]
        self.nut_bang = [[None for _ in range(n)] for _ in range(n)]
        self.tro_choi_ket_thuc_flag = False

        self.tao_giao_dien()

    def tao_giao_dien(self):
        tieude = f"Luật: {self.so_o_thang} ô thẳng hàng"
        tk.Label(self.window, text=tieude, font=('Arial', 14, 'bold'), bg="#f5f5f5").pack(pady=5)

        khung_bang = tk.Frame(self.window, bg="#bbb", bd=1)
        khung_bang.pack(pady=10, padx=10)

        # Điều chỉnh kích thước hiển thị linh hoạt
        size = 15 if self.n > 10 else 25
        font_size = 10 if self.n > 10 else 16

        for i in range(self.n):
            for j in range(self.n):
                nut = tk.Button(khung_bang, text="", font=('Arial', font_size, 'bold'),
                                width=2, height=1, bg="white", relief="flat",
                                command=lambda r=i, c=j: self.xu_ly_nut_click(r, c))
                nut.grid(row=i, column=j, padx=1, pady=1)
                self.nut_bang[i][j] = nut

        self.lbl_status = tk.Label(self.window, text=f"Lượt của: {self.nguoi_choi_hien_tai}", font=('Arial', 12))
        self.lbl_status.pack(pady=5)

        tk.Button(self.window, text="Chơi lại", bg="#e74c3c", fg="white", command=self.reset_game).pack(pady=5)

    def xu_ly_nut_click(self, r, c):
        if self.tro_choi_ket_thuc_flag or self.bang[r][c] != "":
            return

        self.bang[r][c] = self.nguoi_choi_hien_tai
        color = "#d32f2f" if self.nguoi_choi_hien_tai == "X" else "#1976d2"
        self.nut_bang[r][c].config(text=self.nguoi_choi_hien_tai, fg=color, state=tk.DISABLED, disabledforeground=color)

        thang, cac_o_thang = self.kiem_tra_thang_cuoc(r, c)

        if thang:
            self.tro_choi_ket_thuc_flag = True
            for (row, col) in cac_o_thang:
                self.nut_bang[row][col].config(bg="#fff176") # Highlight thắng
            messagebox.showinfo("Kết quả", f"Người chơi {self.nguoi_choi_hien_tai} đã thắng với {self.so_o_thang} ô!")
        elif all(cell != "" for row in self.bang for cell in row):
            messagebox.showinfo("Kết quả", "Hòa!")
        else:
            self.nguoi_choi_hien_tai = "O" if self.nguoi_choi_hien_tai == "X" else "X"
            self.lbl_status.config(text=f"Lượt của: {self.nguoi_choi_hien_tai}")

    def kiem_tra_thang_cuoc(self, r, c):
        huong = [(0, 1), (1, 0), (1, 1), (1, -1)] # Ngang, Dọc, Chéo xuôi, Chéo ngược

        for dr, dc in huong:
            cac_o_thang = [(r, c)]
            # Kiểm tra hai phía của ô vừa đánh
            for huong_di in [1, -1]:
                step = 1
                while True:
                    nr, nc = r + dr * step * huong_di, c + dc * step * huong_di
                    if 0 <= nr < self.n and 0 <= nc < self.n and self.bang[nr][nc] == self.nguoi_choi_hien_tai:
                        cac_o_thang.append((nr, nc))
                        step += 1
                    else:
                        break

            if len(cac_o_thang) >= self.so_o_thang:
                return True, cac_o_thang
        return False, []

    def reset_game(self):
        self.nguoi_choi_hien_tai = "X"
        self.tro_choi_ket_thuc_flag = False
        self.lbl_status.config(text=f"Lượt của: {self.nguoi_choi_hien_tai}")
        for i in range(self.n):
            for j in range(self.n):
                self.bang[i][j] = ""
                self.nut_bang[i][j].config(text="", state=tk.NORMAL, bg="white")

    def run(self):
        self.window.mainloop()

if __name__ == "__main__":
    root = tk.Tk()
    root.withdraw()
    n_input = simpledialog.askinteger("Cài đặt", "Nhập kích thước bàn cờ N:", minvalue=3, maxvalue=20)
    root.destroy()

    if n_input:
        game = CaroNxN(n_input)
        game.run()
```
