```python
#Bài 1:
import copy
import math
import random
import numpy

X = "X"
O = "O"
TRONG =None
nguoi_choi = None
may = None

def trang_thai_ban_dau():
    """
    Returns starting state of the board.
    """
    return [[TRONG, TRONG, TRONG],
            [TRONG, TRONG, TRONG],
            [TRONG, TRONG, TRONG]]

def nguoi_choi_hien_tai(bang):
    """
    Returns player who has the next turn on a board.
    """
    dem = 0
    for i in bang:
        for j in i:
            if j:
                dem += 1
    if dem % 2 != 0:
        return may
    return nguoi_choi

def hanh_dong_kha_thi(bang):
    """
    Returns set of all possible actions (i, j) available on the board.
    """
    ket_qua = set()
    do_dai_bang = len(bang)
    for i in range(do_dai_bang):
        for j in range(do_dai_bang):
            if bang[i][j] == TRONG:
                ket_qua.add((i, j))
    return ket_qua

def ket_qua_nuoc_di(bang, hanh_dong):
    """
    Returns the board that results from making move (i, j) on the board.
    """
    nguoi_choi_dang_hien_hanh = nguoi_choi_hien_tai(bang)
    bang_ket_qua = copy.deepcopy(bang)
    (i, j) = hanh_dong
    bang_ket_qua[i][j] = nguoi_choi_dang_hien_hanh
    return bang_ket_qua

def lay_nguoi_thang_ngang(bang):

    gia_tri_nguoi_thang = None
    do_dai_bang = len(bang)
    for i in range(do_dai_bang):
        gia_tri_nguoi_thang = bang[i][0]
        for j in range(do_dai_bang):
            if bang[i][j] != gia_tri_nguoi_thang:
                gia_tri_nguoi_thang = None
        if gia_tri_nguoi_thang: # Dùng gia_tri_nguoi_thang
            return gia_tri_nguoi_thang # Dùng gia_tri_nguoi_thang
    return gia_tri_nguoi_thang # Dùng gia_tri_nguoi_thang

def lay_nguoi_thang_doc(bang): # Đổi từ get_vertical_winner, board
    # kiểm tra theo chiều dọc # Đổi comment
    gia_tri_nguoi_thang = None # Đổi từ winner_val
    do_dai_bang = len(bang) # Đổi từ board_len
    for i in range(do_dai_bang): # Dùng do_dai_bang
        gia_tri_nguoi_thang = bang[0][i] # Dùng gia_tri_nguoi_thang
        for j in range(do_dai_bang): # Dùng do_dai_bang
            if bang[j][i] != gia_tri_nguoi_thang: # Dùng gia_tri_nguoi_thang
                gia_tri_nguoi_thang = None # Dùng gia_tri_nguoi_thang
        if gia_tri_nguoi_thang: # Dùng gia_tri_nguoi_thang
            return gia_tri_nguoi_thang # Dùng gia_tri_nguoi_thang
    return gia_tri_nguoi_thang # Dùng gia_tri_nguoi_thang

def lay_nguoi_thang_cheo(bang): # Đổi từ get_diagonal_winner, board
    # kiểm tra theo đường chéo # Đổi comment
    gia_tri_nguoi_thang = None # Đổi từ winner_val
    do_dai_bang = len(bang) # Đổi từ board_len
    gia_tri_nguoi_thang = bang[0][0] # Dùng gia_tri_nguoi_thang
    for i in range(do_dai_bang): # Dùng do_dai_bang
        if bang[i][i] != gia_tri_nguoi_thang: # Dùng gia_tri_nguoi_thang
            gia_tri_nguoi_thang = None # Dùng gia_tri_nguoi_thang
    if gia_tri_nguoi_thang: # Dùng gia_tri_nguoi_thang
        return gia_tri_nguoi_thang # Dùng gia_tri_nguoi_thang

    gia_tri_nguoi_thang = bang[0][do_dai_bang - 1] # Dùng gia_tri_nguoi_thang, do_dai_bang
    for i in range(do_dai_bang): # Dùng do_dai_bang
        j = do_dai_bang - 1 - i # Dùng do_dai_bang
        if bang[i][j] != gia_tri_nguoi_thang: # Dùng gia_tri_nguoi_thang
            gia_tri_nguoi_thang = None # Dùng gia_tri_nguoi_thang
    return gia_tri_nguoi_thang # Dùng gia_tri_nguoi_thang

def nguoi_thang(bang):
    """
    Returns the winner of the game, if there is one.
    """
    gia_tri_nguoi_thang = lay_nguoi_thang_ngang(bang) or lay_nguoi_thang_doc(bang) or lay_nguoi_thang_cheo(bang) or None # Dùng tên hàm đã dịch
    return gia_tri_nguoi_thang # Dùng gia_tri_nguoi_thang

def ket_thuc_tro_choi(bang): # Đổi từ terminal, board
    """
    Returns True if game is over, False otherwise.
    """
    if nguoi_thang(bang) != None: # Dùng nguoi_thang
        return True
    for i in bang: # Dùng bang
        for j in i:
            if j == TRONG:
                return False
    return True

def gia_tri_ket_qua(bang): # Đổi từ utility, board
    """
    Returns 1 if X has won the game, -1 if O has won, 0 otherwise.
    """
    gia_tri_nguoi_thang = nguoi_thang(bang) # Dùng nguoi_thang
    if gia_tri_nguoi_thang == X:
        return 1
    elif gia_tri_nguoi_thang == O:
        return -1
    return 0

def gia_tri_toi_da(trang_thai): # Đổi từ maxValue, state
    if ket_thuc_tro_choi(trang_thai): # Dùng ket_thuc_tro_choi
        return gia_tri_ket_qua(trang_thai) # Dùng gia_tri_ket_qua
    v = -math.inf
    for hanh_dong in hanh_dong_kha_thi(trang_thai): # Dùng hanh_dong_kha_thi
        v = max(v, gia_tri_toi_thieu(ket_qua_nuoc_di(trang_thai, hanh_dong))) # Dùng gia_tri_toi_thieu, ket_qua_nuoc_di
    return v

def gia_tri_toi_thieu(trang_thai): # Đổi từ minValue, state
    if ket_thuc_tro_choi(trang_thai): # Dùng ket_thuc_tro_choi
        return gia_tri_ket_qua(trang_thai) # Dùng gia_tri_ket_qua
    v = math.inf
    for hanh_dong in hanh_dong_kha_thi(trang_thai): # Dùng hanh_dong_kha_thi
        v = min(v, gia_tri_toi_da(ket_qua_nuoc_di(trang_thai, hanh_dong))) # Dùng gia_tri_toi_da, ket_qua_nuoc_di
    return v

def minimax(bang): # board thành bang
    """
    Returns the optimal action for the current player on the board.
    """
    nguoi_choi_hien_hanh = nguoi_choi_hien_tai(bang) # Đổi từ current_player
    if nguoi_choi_hien_hanh == X:
        gia_tri_min = -math.inf # Đổi từ min_val
        for hanh_dong in hanh_dong_kha_thi(bang): # Dùng hanh_dong_kha_thi
            kiem_tra = gia_tri_toi_thieu(ket_qua_nuoc_di(bang, hanh_dong)) # Đổi từ check, dùng hàm đã dịch
            if kiem_tra > gia_tri_min: # Dùng kiem_tra, gia_tri_min
                gia_tri_min = kiem_tra # Dùng gia_tri_min, kiem_tra
                nuoc_di = hanh_dong # Đổi từ move
    else:
        gia_tri_max = math.inf # Đổi từ max_val
        for hanh_dong in hanh_dong_kha_thi(bang): # Dùng hanh_dong_kha_thi
            kiem_tra = gia_tri_toi_da(ket_qua_nuoc_di(bang, hanh_dong)) # Đổi từ check, dùng hàm đã dịch
            if kiem_tra < gia_tri_max: # Dùng kiem_tra, gia_tri_max
                gia_tri_max = kiem_tra
                nuoc_di = hanh_dong
    return nuoc_di

if __name__ == "__main__":
    bang = trang_thai_ban_dau()
    luot_may = False
    print("Chọn bên chơi (X hoặc O):")
    nguoi_choi = input()
    if nguoi_choi == "X":
        may = "O"
    else:
        may = "X"

    while True:
        tro_choi_ket_thuc = ket_thuc_tro_choi(bang)
        nguoi_choi_hien_tai_game = nguoi_choi_hien_tai(bang)

        if tro_choi_ket_thuc:
            gia_tri_nguoi_thang = nguoi_thang(bang)
            if gia_tri_nguoi_thang is None:
                print("Trò chơi kết thúc:Hòa gòi cục dàng oiiiii.")
            else:
                print(f"Trò chơi kết thúc:Bé yêu thắng được nhiu đây nè: {gia_tri_nguoi_thang}.")
            break
        else:
            if nguoi_choi != nguoi_choi_hien_tai_game and not tro_choi_ket_thuc:
                if luot_may:
                    nuoc_di = minimax(bang)
                    bang = ket_qua_nuoc_di(bang, nuoc_di)
                    luot_may = False
                print(numpy.array(bang))
            elif nguoi_choi == nguoi_choi_hien_tai_game and not tro_choi_ket_thuc:
                luot_may = True
                print("Chọn vị trí di chuyển nho cưng (hàng, cột từ 0 đến 2):")
                while True:
                    try:
                        i = int(input("Chọn Hàng nha bé:"))
                        j = int(input("Chọn Cột nữa nè:"))
                        if 0 <= i < 3 and 0 <= j < 3 and bang[i][j] == TRONG:
                            break
                        else:
                            print("Lỗi: Nước đi không hợp lệ. Vui lòng nhập hàng và cột trong khoảng 0-2 và chọn một ô trống.")
                    except ValueError:
                        print("Lỗi: Đầu vào không hợp lệ. Vui lòng nhập một số.")

                bang = ket_qua_nuoc_di(bang, (i, j))
                print(numpy.array(bang))
```

```python
#Bài 2:
import os, math
def LayNguoiThang(bang):
    if bang[0] == bang[1] and bang[1] == bang[2]:
        return bang[0]
    elif bang[3] == bang[4] and bang[4] == bang[5]:
        return bang[3]
    elif bang[6] == bang[7] and bang[7] == bang[8]:
        return bang[6]
    elif bang[0] == bang[3] and bang[3] == bang[6]:
        return bang[0]
    elif bang[1] == bang[4] and bang[4] == bang[7]:
        return bang[1]
    elif bang[2] == bang[5] and bang[5] == bang[8]:
        return bang[2]
    elif bang[0] == bang[4] and bang[4] == bang[8]:
        return bang[0]
    elif bang[2] == bang[4] and bang[4] == bang[6]:
        return bang[2]
    return None
def InBang(bang):
    os.system('cls' if os.name == 'nt' else 'clear')
    print(f'''
    {bang[0]} | {bang[1]} | {bang[2]}
    {bang[3]} | {bang[4]} | {bang[5]}
    {bang[6]} | {bang[7]} | {bang[8]}
    ''')

def LayOTrong(bang):
    o_trong = list()
    for o in bang:
        if o != "X" and o != "O":
            o_trong.append(o)
    return o_trong

def minimax(vi_tri, do_sau, alpha, beta, la_nguoi_toi_da):
    nguoi_thang = LayNguoiThang(vi_tri)
    if nguoi_thang != None:
        return 10 - do_sau if nguoi_thang == "X" else -10 + do_sau

    if len(LayOTrong(vi_tri)) == 0:
        return 0

    if la_nguoi_toi_da:
        danh_gia_toi_da = -math.inf
        for o in LayOTrong(vi_tri):
            vi_tri[o - 1] = "X"
            danh_gia = minimax(vi_tri, do_sau + 1, alpha, beta, False)
            danh_gia_toi_da = max(danh_gia_toi_da, danh_gia)
            alpha = max(alpha, danh_gia)
            vi_tri[o - 1] = o
            if beta <= alpha:
                break
        return danh_gia_toi_da
    else:
        danh_gia_toi_thieu = +math.inf
        for o in LayOTrong(vi_tri):
            vi_tri[o - 1] = "O"
            danh_gia = minimax(vi_tri, do_sau + 1, alpha, beta, True)
            danh_gia_toi_thieu = min(danh_gia_toi_thieu, danh_gia)
            beta = min(beta, danh_gia)
            vi_tri[o - 1] = o
            if beta <= alpha:
                break
        return danh_gia_toi_thieu

def TimNuocDiTotNhat(vi_tri_hien_tai, AI):
    gia_tri_tot_nhat = -math.inf if AI == "X" else +math.inf
    nuoc_di_tot_nhat = -1
    for o in LayOTrong(vi_tri_hien_tai):
        vi_tri_hien_tai[o - 1] = AI
        gia_tri_nuoc_di = minimax(vi_tri_hien_tai, 0, -math.inf, +math.inf, False if AI == "X" else True)
        vi_tri_hien_tai[o - 1] = o

        if (AI == "X" and gia_tri_nuoc_di > gia_tri_tot_nhat):
            nuoc_di_tot_nhat = o
            gia_tri_tot_nhat = gia_tri_nuoc_di
        elif (AI == "O" and gia_tri_nuoc_di < gia_tri_tot_nhat):
            nuoc_di_tot_nhat = o
            gia_tri_tot_nhat = gia_tri_nuoc_di

    return nuoc_di_tot_nhat

def main():
    nguoi_choi = input("Bạn muốn chơi với quân X hay O? ").strip().upper()
    AI = "O" if nguoi_choi == "X" else "X"
    game_hien_tai = [*range(1, 10)]
    luot_hien_tai = "X"
    dem = 0

    while True:
        if luot_hien_tai == AI:
            o = TimNuocDiTotNhat(game_hien_tai, AI)
            game_hien_tai[o - 1] = AI
            luot_hien_tai = nguoi_choi
        elif luot_hien_tai == nguoi_choi:
            InBang(game_hien_tai)
            while True:
                input_nguoi_dung = int(input("Nhập số ô bạn muốn đi (1-9): ").strip())
                if input_nguoi_dung in game_hien_tai:
                    game_hien_tai[input_nguoi_dung - 1] = nguoi_choi
                    luot_hien_tai = AI
                    break
                else:
                    InBang(game_hien_tai)
                    print("Ô không có sẵn.")

        if LayNguoiThang(game_hien_tai) != None:
            InBang(game_hien_tai)
            print(f"{LayNguoiThang(game_hien_tai)} THẮNG!!!")
            break

        dem += 1
        if LayNguoiThang(game_hien_tai) == None and dem == 9:
            InBang(game_hien_tai)
            print("Hòa.")
            break

if __name__ == "__main__":
    main()
```
