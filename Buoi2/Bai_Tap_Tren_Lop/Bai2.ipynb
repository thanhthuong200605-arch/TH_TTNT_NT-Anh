```
from heapq import heappush, heappop

# ================= NODE =================
class Node:
    def __init__(self, parent, name, g, h):
        self.parent = parent   # node cha
        self.name = name       # tên đỉnh
        self.g = g             # chi phí từ start
        self.h = h             # heuristic

    def __lt__(self, other):
        return (self.g + self.h) < (other.g + other.h)


# ================= IN ĐƯỜNG ĐI =================
def print_path(node):
    path = []
    while node:
        path.append(node.name)
        node = node.parent
    path.reverse()
    print("Đường đi:", " -> ".join(path))


# ================= A* =================
def A_star(graph, heuristic, start, goal):
    open_list = []   # priority queue
    visited = set()  # tập đã duyệt

    root = Node(None, start, 0, heuristic[start])
    heappush(open_list, root)

    while open_list:
        current = heappop(open_list)

        # Nếu đã xét rồi thì bỏ qua
        if current.name in visited:
            continue
        visited.add(current.name)

        # Nếu tới đích
        if current.name == goal:
            print("Tìm thấy đường đi!")
            print_path(current)
            print("Chi phí:", current.g)
            return

        # Duyệt các đỉnh kề
        for neighbor, cost in graph[current.name]:
            if neighbor not in visited:
                child = Node(
                    current,
                    neighbor,
                    current.g + cost,
                    heuristic[neighbor]
                )
                heappush(open_list, child)

    print("Không tìm thấy đường đi")


# ================= MAIN TEST =================
if __name__ == "__main__":

    # Đồ thị (adjacency list)
    graph = {
        'A': [('B', 1), ('C', 3)],
        'B': [('D', 3), ('E', 1)],
        'C': [('F', 5)],
        'D': [],
        'E': [('D', 1), ('F', 2)],
        'F': []
    }

    # Heuristic (ước lượng tới F)
    heuristic = {
        'A': 6,
        'B': 4,
        'C': 4,
        'D': 1,
        'E': 2,
        'F': 0
    }

    start = 'A'
    goal = 'F'

    A_star(graph, heuristic, start, goal)
```
